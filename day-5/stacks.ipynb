```python
LOG = True
def log(txt):
    if LOG:
        print(txt)
```

```python
import re

def move_crate(stack_from, stack_to):
    crate = stack_from.pop()
    log(f' -moving {crate} from {stack_from} to {stack_to}-')
    stack_to.append(crate)

def move_multiple_crates(stack_from, stack_to, amount):
    crates = stack_from[-amount:]
    del stack_from[-amount:]
    log(f' -moving {crates} from {stack_from} to {stack_to}-')
    stack_to += crates

def display_stacks(stacks):
    for index, stack in enumerate(stacks):
        string = f'{index} | '
        for item in stack:
            string += f'[{item}]'
        log(string)


with open('input.txt') as f:
    data = f.read()

split = data.split('\n\n')
stacks_input = split[0]
instructions = split[1]
log('---')
log(stacks_input)
log('---')
log(instructions)
log('---')

split_stacks = stacks_input.split('\n')
# log(split_stacks)
num_stacks = (len(split_stacks[-1])+1)/4
# log(f'num stacks: {num_stacks}')

stacks = [[] for _ in range(int(num_stacks))]
# log(stacks)

for n in reversed(range(len(split_stacks)-1)):
    # log(f'n: {n}')
    for i in range(int(num_stacks)):
        crate = split_stacks[n][1+i*4]
        # log(f' crate: {crate}')
        if crate != ' ':
            stacks[i].append(crate)

log(f'stacks: {stacks}')

split_instructions = instructions.split('\n')

for instruction in split_instructions:

    display_stacks(stacks)

    # log(f' instruction: {instruction}')
    # move_amount = int(instruction[re.search('move ', instruction).span()[1]])
    # move_from = int(instruction[re.search('from ', instruction).span()[1]])
    # move_to = int(instruction[re.search('to ', instruction).span()[1]])
    move_amount, move_from, move_to = [int(char) for char in re.findall(r'\d+', instruction)]
    log(f' amount: {move_amount} from: {move_from-1} to: {move_to-1}')
    move_multiple_crates(stacks[move_from-1], stacks[move_to-1], move_amount)
    # for i in range(move_amount):
        # move_crate(stacks[move_from-1], stacks[move_to-1])

display_stacks(stacks)
    
# Get the top crates
msg = ''
for index, stack in enumerate(stacks):
    if len(stack) > 0:
        msg += stack[-1]
print(f'msg: {msg}')
```
